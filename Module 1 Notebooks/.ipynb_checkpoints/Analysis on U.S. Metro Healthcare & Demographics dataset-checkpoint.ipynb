# Visualizing relationship between features

```python
#Relation Between Physicians and hospital bed
plt.figure(figsize=(14,6))
plt.scatter(df['NumMDs'], df['NumBeds'], c='blue', alpha=0.5)  # Adjust colors as needed
plt.xlabel('Number of Physicians')
plt.ylabel('Number of Hospital Beds')
plt.title('Relationship between Physicians and Hospital Beds')
plt.scatter(df['NumMDs'], df['NumBeds'], c=df['NumMedicare'], s=df['RateBeds'], alpha=0.5, cmap='coolwarm')
plt.colorbar(label='Number of Medicare recipients')
plt.xscale('log')
plt.grid(True)
plt.show()
```

```python
#this visulization give inshght of different hospital condition,like anlyse how many beds are there so ,help in and disastrous time
correlation_coefficient = df['NumMDs'].corr(df['NumBeds'])
if correlation_coefficient > 0:
    strength = 'strong positive'
elif correlation_coefficient < 0:
    strength = 'strong negative'
else:
    strength = 'no'

print("Correlation Coefficient:", correlation_coefficient)
print("There is", strength, "correlation between the number of physicians and the number of hospital beds.")
```

```python
#Visualize Number of doctors and hospital accros each city
import matplotlib.pyplot as plt

# Bar chart for number of doctors for each metropolitan area
plt.figure(figsize=(12, 10))
plt.bar(df['City'], df['NumMDs'], color='skyblue', edgecolor='black', linewidth=1, alpha=0.8)
plt.xlabel('Metropolitan Area', fontsize=12)
plt.ylabel('Number of Doctors', fontsize=12)
plt.title('Number of Doctors for Each Metropolitan Area', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# Bar chart for number of hospitals for each metropolitan area
plt.figure(figsize=(12, 10))
plt.bar(df['City'], df['NumHospitals'], color='lightgreen', edgecolor='black', linewidth=1, alpha=0.8)
plt.xlabel('Metropolitan Area', fontsize=12)
plt.ylabel('Number of Hospitals', fontsize=12)
plt.title('Number of Hospitals for Each Metropolitan Area', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# Doctor-Hospital Ratio calculation
df['Doctor_Hospital_Ratio'] = df['NumMDs'] / df['NumHospitals']

# Sorting the dataset by Doctor-Hospital Ratio
df_sorted = df.sort_values(by='Doctor_Hospital_Ratio', ascending=False)

# Bar chart for Doctor-Hospital Ratio for each metropolitan area
plt.figure(figsize=(12, 8))
plt.bar(df_sorted['City'], df_sorted['Doctor_Hospital_Ratio'], color='blue', edgecolor='black', linewidth=1)
plt.xlabel('Metropolitan Area', fontsize=12)
plt.ylabel('Doctor-Hospital Ratio', fontsize=12)
plt.title('Doctor-Hospital Ratio for Each Metropolitan Area', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
#Relationship between Number of Hospitals and Number of Doctors: Scatter Plot with Regression Lines

# Scatter plot with regression lines
sns.regplot(data=df, x='NumHospitals', y='NumMDs', scatter_kws={'color': 'blue', 'alpha': 0.6}, line_kws={'color': 'red'})
plt.xlabel('Number of Hospitals', fontsize=12)
plt.ylabel('Number of Doctors', fontsize=12)
plt.title('Relationship between Number of Hospitals and Number of Doctors', fontsize=14, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# Show Best and need to improve areas in health sector 

```python
#Comparison of Doctor-Hospital Ratios: Top Areas with the Best Healthcare Infrastructure vs. 
#Areas in Need of Further Improvement
# Doctor-Hospital Ratio calculation
df['Doctor_Hospital_Ratio'] = df['NumMDs'] / df['NumHospitals']

# Sorting the dataset by Doctor-Hospital Ratio
df_sorted = df.sort_values(by='Doctor_Hospital_Ratio', ascending=False)

# Selecting top areas with the best healthcare infrastructure
top_areas = df_sorted.head(5)

# Selecting bottom areas in need of further improvement
bottom_areas = df_sorted.tail(5)

# Bar chart for top areas with the best healthcare infrastructure
plt.figure(figsize=(10, 6))
plt.bar(top_areas['City'], top_areas['Doctor_Hospital_Ratio'], color='green', edgecolor='black', linewidth=1)
plt.xlabel('Metropolitan Area', fontsize=12)
plt.ylabel('Doctor-Hospital Ratio', fontsize=12)
plt.title('Top Areas with the Best Healthcare Infrastructure', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# Bar chart for bottom areas in need of further improvement
plt.figure(figsize=(10, 6))
plt.bar(bottom_areas['City'], bottom_areas['Doctor_Hospital_Ratio'], color='red', edgecolor='black', linewidth=1)
plt.xlabel('Metropolitan Area', fontsize=12)
plt.ylabel('Doctor-Hospital Ratio', fontsize=12)
plt.title('Bottom Areas in Need of Further Improvement', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# Create the report
report = f"""
Healthcare Services Analysis Report

Top areas with the best healthcare infrastructure:
{top_areas[['City', 'Doctor_Hospital_Ratio']].to_string(index=False)}

Bottom areas in need of further improvement:
{bottom_areas[['City', 'Doctor_Hospital_Ratio']].to_string(index=False)}
"""

# Print the report
print(report)
```

# Model Trainig linear regression

```python
x = df[['NumHospitals']]
```

```python
x.shape
```

```python
y=df[['MedicareRate']]
```

```python
#split
#Splitting the datset into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,
                                                  random_state=42,shuffle=True)
```

```python
#model training
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```
